# Processing the Output

What output will be saved can be spefified in `rtmf6.toml`:

```toml
[output]
equilibrium_phases = "out/equilibrium_phases"
concentrations = "out/concentrations"
selected_output = "out/selected_output"
```

The values are stored in Python `shelve` files.
You can read them with the `ShelveViewer`.
Import it:

````python
from rtmf6.postprocessing.output_reader import ShelveViewer
````

Make an instance, provding the path to the config file:

````python
viewer = ShelveViewer('rtmf6.toml')
````

The viewer provides several attributes:

````python
viewer
````

Each attribute alos offers attributes:

````python
viewer.concentrations
````

The Aluminun contration is aviailbale for severals time steps:

````python
al = viewer.concentrations.Al
al
````

````python
al.time_steps[:5]
````

````python
last_step = al.time_steps[-1]
last_step
````

## Selected Output

The selected out can be defined in the file specified with `chemistry_name` in the config file:

```toml
[phreeqcrm]
number_of_threads = 1
chemistry_name = "phreeqcrm/advect.pqi"
```

For example:

```
SELECTED_OUTPUT
        -file            advect.sel
        -reset           false
        -pH              true
        -pe              true
        -totals          Na Cl K Ca
END
```

The results of recorded base on this are in  `selected_output_1`:

````python
viewer.selected_output
````

````python
viewer.selected_output.selected_output_1
````

It is a dictionary of all names:

````python
select_last = viewer.selected_output.selected_output_1.get_value(last_step)
select_last.keys()
````

This is the pH value:

````python
ph = select_last['pH']
````

````python
ph.shape
````

````python
ph.min()
````

````python
ph.max()
````

````python
ph
````

````python
select_last_333 = viewer.selected_output.selected_output_333.get_value(last_step)
print(*select_last_333.keys(), sep='\n')
````
